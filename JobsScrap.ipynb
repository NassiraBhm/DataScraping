```python
from bs4 import BeautifulSoup
import requests
import re
import time
```

```python
def job_search():
    url = 'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=&searchTextText=&txtKeywords=data+analyst&txtLocation=&cboWorkExp1=-1#'
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser').prettify()
    soup = BeautifulSoup(soup, 'html.parser')

    jobs = soup.find_all('li', class_= 'clearfix job-bx wht-shd-bx')
    for index, job in enumerate(jobs):
       posted_date = job.find('span', class_ = 'sim-posted').span.text 
       if 'few' in posted_date:
          companies = job.find('h3', class_ = 'joblist-comp-name').text.replace(' ', '')
          skills = job.find('span', class_ = 'srp-skills').text.replace("\n", "")
          skills = re.sub(r"\s+", " ", skills)
          Description = job.header.h2.a['href']
          index += 1
          with open('post{}'.format(index) + '.txt', 'w') as f:
            f.write(f'''Company name: {companies.strip()}\n'''  )
            f.write(f'''Required skills: {skills.strip()}\n''' )   
            f.write(f'''Job description: {Description}''') 
            print(' ')

if __name__ == '__main__':
    while True:
        job_search()
        print('Waiting 1 hour')
        time.sleep(3600)
```
